```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno


np.random.seed(0)
data = pd.read_csv('airbnb.csv')
df = pd.DataFrame(data)


msno.matrix(df)
plt.show()


df_missing = df.isnull().melt(id_vars=['rating'], var_name='column', value_name='is_missing')


missing_by_rating = df_missing.groupby(['rating', 'column'])['is_missing'].mean().reset_index()


plt.figure(figsize=(12, 6))
sns.barplot(x='rating', y='is_missing', hue='column', data=missing_by_rating)
plt.title('Missingness Based on Rating')
plt.xlabel('Rating')
plt.ylabel('Proportion of Missing Values')
plt.show()

```

```python
# Buat dalam bentuk barplot
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(0)
np.random.seed(0)
data = pd.read_csv('airbnb.csv')
df = pd.DataFrame(data)


df_missing = df.isnull().melt(id_vars=['rating'], var_name='column', value_name='is_missing')


missing_by_rating = df_missing.groupby(['rating', 'column'])['is_missing'].mean().reset_index()


plt.figure(figsize=(12, 6))
sns.barplot(x='rating', y='is_missing', hue='column', data=missing_by_rating)
plt.title('Proportion of Missing Values by Rating')
plt.xlabel('Rating')
plt.ylabel('Proportion of Missing Values')
plt.legend(title='Column', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Cek DataFrame yang ada missing values di kolom rating, number_of_stays, 5_stars, reviews_per_month
import pandas as pd
import numpy as np

df = pd.DataFrame(data)

print("DataFrame Asli:")
print(df)

missing_columns = ['rating', 'number_of_stays', '5_stars', 'reviews_per_month']
df_missing = df[df[missing_columns].isnull().any(axis=1)]

print("\nDataFrame dengan Missing Values di Kolom Tertentu:")
print(df_missing)
```

```python
# Lakukan imputasi untuk missing data
df = pd.DataFrame(data)

print("DataFrame Asli:")
print(df)

df_mean_imputed = df.fillna(df.mean())
print("\nDataFrame dengan Imputasi Mean:")
print(df_mean_imputed)

df_median_imputed = df.fillna(df.median())
print("\nDataFrame dengan Imputasi Median:")
print(df_median_imputed)

df_mode_imputed = df.copy()
for column in df_mode_imputed.columns:
    df_mode_imputed[column].fillna(df_mode_imputed[column].mode()[0], inplace=True)
print("\nDataFrame dengan Imputasi Modus:")
print(df_mode_imputed)
```
